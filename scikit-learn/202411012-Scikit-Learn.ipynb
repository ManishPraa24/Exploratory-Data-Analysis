```python
# Importing required libraries

import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Importing dataset

url = r'https://raw.githubusercontent.com/ManishPraa24/Exploratory-Data-Analysis/refs/heads/main/Datasets/kamyr-digester.csv'

data = pd.read_csv(url)

data
```

```python
# Checking for data-columns

attributes = data.columns

print(attributes)
```

<h2>
    1. Pre-processing Data
</h2>

```python
# Checking for null values

data.isnull().sum()
```

```python
data.dtypes
```

<h3>
    Here, all my data type is float. Hence, I will use SimpleImputer for filling the values.
</h3>

```python
attributes = list(data.columns)
print(attributes)

print()

del attributes[0]
print(attributes)
```

```python
# Using simple imputer

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])
```

```python
# Preprocessing the data

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, attributes),
    ])
```

```python
data_transformed = preprocessor.fit_transform(data)
```

```python
data_transformed
```

```python
# Converting it to data Frames

data_transformed = pd.DataFrame(data_transformed, columns=attributes)

data_transformed
```

```python
# Checking for null values in the newly created data frame

data_transformed.isnull().sum()
```

<h2>
    2. Training a regression model
</h2>

<h3>
    Here, we will take Y-kappa as our target variable
</h3>

```python
# Splitting the data into training and testing

X = data_transformed.drop('Y-Kappa', axis=1)
y = data_transformed['Y-Kappa']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)
```

```python
# Initializing the random forest model

rf_model = RandomForestRegressor(random_state=24)
```

```python
# Fitting the model

rf_model.fit(X_train, y_train)
```

```python
# Predicting the y-kappa

y_pred = rf_model.predict(X_test)
```

<h2>
    3. Model Evaluation
</h2>

```python
# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("R² Score:", r2)
```

<h2>
    4. Exporting Model
</h2>

```python
# Save the model to a file
with open('trained_model.pkl', 'wb') as file:
    pickle.dump(rf_model, file)
```
