<h3>
    3. MCAR Test
</h3>
<br>

```python
# Drop columns with more than 50% missing values
missing_rate = numeric_data.isnull().mean()
print(missing_rate)
df_filtered = numeric_data.loc[:, missing_rate < 0.5]
df_filtered
```

```python
# Check if there's any numeric data to process
if df_filtered.empty:
    raise ValueError("No numeric columns with less than 50% missing data found in the dataset. Little's MCAR test cannot be performed.")
```

```python
# Creating a function for performing little MCAR test

def little_mcar_test(data):

    # Convert dataframe to numpy array
    data = data.to_numpy()

    # Count missing values
    missing_values = np.isnan(data)

    # If no missing values, return early
    if not np.any(missing_values):
        print("No missing values found. Data is MCAR by definition.")
        return

    # Calculate missing data pattern
    missing_patterns = np.sum(missing_values, axis=1)

    # Calculate statistics for missing data pattern
    unique_patterns, counts = np.unique(missing_patterns, return_counts=True)
    pattern_probabilities = counts / len(missing_patterns)

    # Perform chi-square test to check the pattern distribution
    chi2_stat, p_value = stats.chisquare(pattern_probabilities)

    return chi2_stat, p_value
    
```

```python
# Creating a function to interpret the results of MCAR test

def interpretResult(chi_square, p_value):
    print("MCAR Test Results are as follows: \n")
    print(f"Chi-squared statisitcs: {chi_square}")
    print(f"p_value: {p_value}")
    if p_value > 0.05:
        print("The null hypothesis cannot be rejected. The data is MCAR.")
    else:
        print("The null hypothesis is rejected. The data is not MCAR.")
        
```

```python
chi_squared, p_value = little_mcar_test(df_filtered)

print(f"Chi-square statistic: {chi_squared:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Interpreting the chi_square and p_value


interpretResult(chi_squared, p_value)
```
